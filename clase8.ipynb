# EJEMPLOS DE GRAFOS CON NETWORK

# Creaci√≥n de grafo

```python
import networkx as nx 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
G= nx.Graph()

G.add_edge('v1','v2', weight=10)
G.add_edge('v2','v3', weight=5)
G.add_edge('v1','v4', weight=9)
G.add_edge('v2','v4', weight=6)
G.add_edge('v4','v5', weight=5)
G.add_edge('v3','v5', weight=7)
G.add_edge('v3','v6', weight=9)
G.add_edge('v5','v6', weight=6)

pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=500)

nx.draw_networkx_edges(G, pos, width=20)

nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
nx.draw_networkx_edge_labels(G, pos)

plt.axis('off')
plt.show()
```

# C√°lculo de la densidad 

```python
nx.density(G)
```

# Vecinos nodo 3

```python
for p in nx.all_neighbors(G,'v3'):
    print(p)
```

# Matriz de adyacencia

```python
a=nx.adjacency_matrix(G)
a.shape
```

```python
b=nx.incidence_matrix(G)
b.toarray()
```

```python
c=nx.laplacian_matrix(G)
c.toarray()
```
